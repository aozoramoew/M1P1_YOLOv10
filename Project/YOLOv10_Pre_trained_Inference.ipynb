## **4. Create YOLO model**

```python
from ultralytics import YOLOv10

MODEL_PATH = 'yolov10n.pt'
model = YOLOv10(MODEL_PATH)
```

```python
model.info()
```

## **5. Run prediction with an image**

```python
# https://github.com/googlecolab/colabtools/issues/3409
import locale
locale.getpreferredencoding = lambda: "UTF-8" #Preferred Encoding = UTF-8
```

```python
from google.colab.patches import cv2_imshow #To display image in GG Colab

IMAGE_URL = 'https://www.tripsavvy.com/thmb/z1lmhJxYMQ6ScsISa3qHzeArsL0=/1500x0/filters:no_upscale():max_bytes(150000):strip_icc()/street-view-of-pham-ngu-lao-street--vietnam-921731936-5c353f9746e0fb000180251c.jpg'
CONF_THRESHOLD = 0.5 #Only Objects detected with a confidence score of 0.5 or higher will be considered
IMG_SIZE = 640 #Defines the size to which the input image will be resized before being fed into the model
results = model.predict(source=IMAGE_URL,
                       imgsz=IMG_SIZE,
                       conf=CONF_THRESHOLD)
annotated_img = results[0].plot()

cv2_imshow(annotated_img)
```
